```python
# Cell 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cell 2: Load CSV File
df_csv = pd.read_csv('cars.csv')
print('CSV Data Loaded ')
df_csv.head()
```

```python
# Cell 3: Load Excel File
df_excel = pd.read_excel('cars.xlsx')
print('Excel Data Loaded ✅')
df_excel.head()
```

```python
# Cell 4: Load from Online Repository
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv'
df_online = pd.read_csv(url)
print('Online Repo Data Loaded ✅')
df_online.head()
```

```python
# Cell 5: Basic DataFrame Operations
print('CSV Shape:', df_csv.shape)
print('Excel Shape:', df_excel.shape)
print('Online Shape:', df_online.shape)

print('Online Dataset Info:')
df_online.info()

print('Descriptive Statistics (Online Dataset):')
df_online.describe(include='all')

print('Missing Values (Online Dataset):', df_online.isnull().sum())
```

```python
# Cell 6: Preprocessing
df_online_clean = df_online.dropna()  # Drop missing values
print('Cleaned Online DataFrame ✅')
df_online_clean.head()
```

```python
# Cell 7: Statistical Analysis
corr = df_online_clean.corr(numeric_only=True)
print('Correlation Matrix:')
print(corr)

print('Summary Statistics:')
print(df_online_clean.describe())
```

```python
# Cell 8: Visualization
plt.figure(figsize=(8,5))
plt.hist(df_online_clean['mpg'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(df_online_clean['horsepower'], df_online_clean['mpg'], alpha=0.6, color='orange', edgecolor='black')
plt.title('Horsepower vs MPG')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()

plt.figure(figsize=(6,5))
plt.boxplot(df_online_clean['mpg'])
plt.title('Boxplot of MPG')
plt.ylabel('MPG')
plt.show()
```

```python
# Cell 9: Create Small Custom Dataset
small_data = {
    'Name': ['Car A','Car B','Car C','Car D','Car E','Car F','Car G'],
    'Horsepower': [100,150,200,120,180,160,140],
    'MPG': [25,22,18,28,20,21,24],
    'Price': [20000,25000,30000,22000,28000,26000,24000]
}
df_small = pd.DataFrame(small_data)
print('Small Custom Dataset ✅')
df_small
```

```python
# Cell 10: Visualize Small Dataset
# Bar chart for Price
plt.figure(figsize=(8,5))
plt.bar(df_small['Name'], df_small['Price'], color='lightgreen')
plt.title('Price of Each Car')
plt.xlabel('Car Name')
plt.ylabel('Price')
plt.show()

# Scatter plot: Horsepower vs MPG
plt.figure(figsize=(8,5))
plt.scatter(df_small['Horsepower'], df_small['MPG'], color='red')
plt.title('Horsepower vs MPG (Small Dataset)')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()
```
